```python
df_all_wine = pd.concat([white_wine_df, red_wine_df])
df_all_wine.rename(columns = {'fixed acidity':'fixed_acidity'}, inplace = True)
df_all_wine.rename(columns = {'volatile acidity':'volatile_acidity'}, inplace = True)
df_all_wine.rename(columns = {'citric acid':'citric_acid'}, inplace = True)
df_all_wine.rename(columns = {'residual sugar':'residual_sugar'}, inplace = True)
df_all_wine.rename(columns = {'free sulfur dioxide':'free_sulfur_dioxide'}, inplace = True)
df_all_wine.rename(columns = {'total sulfur dioxide':'total_sulfur_dioxide'}, inplace = True)
df_all_wine = df_all_wine.sample(frac=1).reset_index(drop=True)
df_all_wine
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df_all_wine, test_size=0.2)
train_df, val_df = train_test_split(train_df, test_size=0.2)
```

```python
import tensorflow as tf

from tensorflow import feature_column
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split

def df_to_dataset(dataframe, shuffle=True, batch_size=32):
    dataframe = dataframe.copy()
    labels = dataframe.pop('target')
    ds = tf.data.Dataset.from_tensor_slices((dict(dataframe), labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(dataframe))
    ds = ds.batch(batch_size)
    return ds
batch_size = 5 # A small batch sized is used for demonstration purposes
train_ds = df_to_dataset(train_df, batch_size=batch_size)
val_ds = df_to_dataset(val_df, shuffle=False, batch_size=batch_size)
#test_ds = df_to_dataset(test_df, shuffle=False, batch_size=batch_size)
```

```python
batch_size = 32
train_ds = df_to_dataset(train_df, batch_size=batch_size)
val_ds = df_to_dataset(val_df, shuffle=False, batch_size=batch_size)
test_ds = df_to_dataset(test_df, shuffle=False, batch_size=batch_size)
```

```python
feature_columns = []

# numeric cols
for header in ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar', 'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']:
    feature_columns.append(feature_column.numeric_column(header))
feature_layer = tf.keras.layers.DenseFeatures(feature_columns)

feature_columns
```

```python

model = tf.keras.Sequential([
  feature_layer,
  layers.Dense(128, activation='relu'),
  layers.Dense(128, activation='relu'),
  layers.Dropout(.1),
  layers.Dense(1)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])

model.fit(train_ds,
          validation_data=val_ds,
          epochs=10)
```

```python
loss, accuracy = model.evaluate(test_ds)
print("Accuracy", accuracy)
test_instance={
"fixed_acidity":6.4000,
"volatile_acidity":0.2600,
"citric_acid":0.4900,
"residual_sugar":6.4000,
"chlorides":0.0370,
"free_sulfur_dioxide":37.0000,
"total_sulfur_dioxide":161.0000,
"density":0.9954,
"pH":3.3800,
"sulphates":0.5300,
"alcohol":9.7000,
"quality":5.0
}

input_dict = {name: tf.convert_to_tensor([value]) for name, value in test_instance.items()}
prediction = model.predict(input_dict)
prob = tf.nn.sigmoid(prediction[0])

print(
    "This particular wine had a %.1f percent probability "
    "being a red wine." % (100 * prob)
)
print(
    "This particular wine had a %.1f percent probability "
    "being a white wine." % (100 * prob)
)
```
